## Question 1: Boolean Operations on Distance Fields (10 points)

Write a code snippet to perform Boolean operations of (i) Union (ii) Intersection and (iii) difference of two distance fields. This task is highlighted by a TODO below.


```python
def Union(obj1, obj2):
    # Union operation for two objects
    # obj1 Union obj2
    # TODO: Union operation 
    #union_array = max(obj1,obj2)
    union_array = np.where(obj1>obj2, obj1, obj2)
    return union_array
```

```python
def Intersection(obj1, obj2):
    # Intersection operation for two objects
    # obj1 Intersection obj2
    # TODO: Intersection operation 
    #int_array = min(obj1,obj2)
    int_array = np.where(obj1<obj2, obj1, obj2)
    return int_array
```

```python

def Difference(obj1, obj2):
    # Difference operation for two objects
    # obj1 Minus obj2
    # TODO: Difference operation
    #diff_array = min(obj1,-obj2)
    diff_array = np.where(obj1<(-1.0*obj2), obj1, (-1.0*obj2))
    return diff_array
```

```python
# Parameters that control the distance fields
gridsize = 0.25
origin = [0.0, 0.0, 0.0]

# The distancefields.npz is a numpy array that contains the distance fields of a cube and a cylinder
cube, cylinder =  load_difis('distancefields.npz')

print(cube)
print(cylinder)
```

```python
# Perform the Boolean operations and store the distance fields in new variables
union = Union(cube, cylinder)
intersection = Intersection(cube, cylinder) 
diff = Difference(cube, cylinder)
```
